# Attrition = No

```python
df = df1.loc[df1['Attrition']=="No"]
df
```

```python
dfhours = df[df.columns[11:13]]
dfhours

dfhours.describe()
```

```python
#EnvironmentSatisfaction
elow = 0
ehigh = 0

nno = len(df[df.columns[10:11]])

e = df[df.columns[10:11]].values

for i in e:
    if 1 <= i <= 2:
        elow += 1
    if 3 <= i <= 4:
        ehigh += 1
    

pelow = round(elow/nno * 100, 2)
pehigh = round(ehigh/nno * 100, 2)
        
print("Number of employees with low Environment Satisfaction: " + str(elow) + " which is " + str(pelow) + " %.")
print("Number of employees with high Environment Satisfaction: " + str(ehigh) + " which is " + str(pehigh) + " %.")
```

```python
#JobInvolvement
jilow = 0
jihigh = 0

ji = df[df.columns[13:14]].values

for i in ji:
    if 1 <= i <= 2:
        jilow += 1
    if 3 <= i <= 4:
        jihigh += 1

pjilow = round(jilow/nno * 100, 2)
pjihigh = round(jihigh/nno * 100, 2)

print("Number of employees with low Job Involvement: " + str(jilow) + " which is " + str(pjilow) + " %.")
print("Number of employees with high Job Involvement: " + str(jihigh) + " which is " + str(pjihigh) + " %.")
```

```python
#Gender
female = 0
male = 0

gender = df[df.columns[11:12]].values

for i in gender:
    if i == "Female":
        female += 1
    if i == "Male":
        male += 1
        
pfemale = round(female/nno * 100,2)
pmale = round(male/nno * 100,2)
        
print("Number of Female employees: " + str(female) + " which is " + str(pfemale) + " %.")
print("Number of Male employees: " + str(male) + " which is " + str(pmale) + " %.")
```

# Attrition = YES
### analysis of the parameters for employees which no longer work for the company

```python
#taking only values with attrition = yes
dfA = df1.loc[df1['Attrition']=="Yes"]
dfA.head()
```

```python
dfhours = dfA[dfA.columns[11:13]]
dfhours

dfhours.describe()
```

```python
#EnvironmentSatisfaction
e1low = 0
e1high = 0
n1 = len(dfA[dfA.columns[10:11]])

e1 = dfA[dfA.columns[10:11]].values

for i in e1:
    if 1 <= i <= 2:
        e1low += 1
    if 3 <= i <= 4:
        e1high += 1

        
pe1low = round(e1low/n1*100,2)
pe1high = round(e1high/n1*100,2)

print("Number of employees who left the company with low Environment Satisfaction: " + str(e1low) + " which is " + str(pe1low) + " %.")
print("Number of employees who left the company with high Environment Satisfaction: " + str(e1high) + " which is " + str(pe1high) + " %.")

```

```python
#JobInvolvement
ji1low = 0
ji1high = 0
n1j = len(dfA[dfA.columns[13:14]])
ji1 = dfA[dfA.columns[13:14]].values

for i in ji1:
    if 1 <= i <= 2:
        ji1low += 1
    if 3 <= i <= 4:
        ji1high += 1

pji1low = round(ji1low/n1j * 100,2)
pji1high = round(ji1high/n1j * 100,2)

print("Number of employees with low Job Involvement: " + str(ji1low) + " which is " + str(pji1low) + " %.")
print("Number of employees with high Job Involvement: " + str(ji1high) + " which is " + str(pji1high ) + " %.")
```

```python
#Gender
female1 = 0
male1 = 0

gender1 = dfA[dfA.columns[11:12]].values

for i in gender1:
    if i == "Female":
        female1 += 1
    if i == "Male":
        male1 += 1
        
pfemale1 = round(female1/n1 * 100,2)
pmale1 = round(male1/n1 * 100,2)       

print("Number of Female employees: " + str(female1) + " which is " + str(pfemale1) + " %.")
print("Number of Male employees: " + str(male1) + " which is " + str(pmale1) + " %.")
```

```python
All = {"EnvironmentSatisfaction": [elow, ehigh], "JobInvolvement": [jilow, jihigh], "Gender": [female, male]}
pAll = {"EnvironmentSatisfaction": [pelow, pehigh], "JobInvolvement": [pjilow, pjihigh], "Gender": [pfemale, pmale]}
Ayes = {"EnvironmentSatisfaction": [e1low, e1high], "JobInvolvement": [ji1low, ji1high], "Gender": [female1, male1]}
pAyes = {"EnvironmentSatisfaction": [pe1low, pe1high], "JobInvolvement": [pji1low, pji1high], "Gender": [pfemale1, pmale1]}
```

```python
dfAll = pd.DataFrame(All, index= ["low", "high"])
dfAll
```

```python
dfpAll = pd.DataFrame(pAll, index= ["low", "high"])
dfpAll
```

```python
dfAyes = pd.DataFrame(Ayes, index= ["low", "high"])
dfAyes
```

```python
dfpAyes = pd.DataFrame(pAyes, index= ["low", "high"])
dfpAyes
```

```python
dftogether = [dfAll, dfpAll, dfAyes, dfpAyes]
result = pd.concat(dftogether, keys=["ANo", "p%No" ,"AYes", "p%Ayes"])
result
```

Results: Enviroment Satisfaction and JobInvolvment are around 12% lower for employees who have left the company.
