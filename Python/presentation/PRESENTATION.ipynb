### Attrition Breakdown

```python
df_no = df.loc[df['Attrition']== "No"]
df_no
```

```python
df_yes = df.loc[df['Attrition']== "Yes"]
df_yes
```

```python
print(df_no["Attrition"].count(),df_yes["Attrition"].count())
```
